## Workflow

```python
import gmns_ready as gr
print(f"Using gmns-ready v{gr.__version__}\n")
```

### Step 1: Validate Spatial Alignment

```python
# Make the current Python process use UTF-8 for stdout/stderr
import os, sys

# Prefer UTF-8 for all I/O in this process
os.environ["PYTHONUTF8"] = "1"
os.environ["PYTHONIOENCODING"] = "utf-8"

# Reconfigure the already-opened streams (Python 3.7+)
try:
    sys.stdout.reconfigure(encoding="utf-8")
    sys.stderr.reconfigure(encoding="utf-8")
except Exception:
    pass
```

```python
gr.validate_basemap()
```

### Step 2: Extract Zones

```python
gr.extract_zones()
```

```python
# Preview zones
import pandas as pd
zones = pd.read_csv('zone.csv')
print(f"Extracted {len(zones)} zones\n")
zones[['zone_id', 'x_coord', 'y_coord']].head()
```

### Step 3: Build Zone-Connected Network

```python
gr.build_network()
```

```python
# Check results
nodes = pd.read_csv('connected_network/node.csv')
links = pd.read_csv('connected_network/link.csv')
connectors = pd.read_csv('connected_network/connector_links.csv')

print(f"ðŸ“Š Network Statistics:")
print(f"  Zones: {len(zones)}")
print(f"  Total nodes: {len(nodes)}")
print(f"  Total links: {len(links)}")
print(f"  Connectors: {len(connectors)}")
```

### Step 4: Validate Network

```python
gr.validate_network()
```

### Step 5: Validate Accessibility

```python
gr.validate_accessibility()
```

```python
# Analyze accessibility
accessibility = pd.read_csv('connected_network/zone_accessibility.csv')

print(f"ðŸ“Š Accessibility Results:")
print(f"  Avg destinations reachable: {accessibility['reachable_destinations'].mean():.1f}")
print(f"  Avg origins connecting: {accessibility['connecting_origins'].mean():.1f}")

# Show poorly connected zones
threshold = accessibility['reachable_destinations'].mean() * 0.5
poor_zones = accessibility[accessibility['reachable_destinations'] < threshold]
print(f"\nâš ï¸  Zones with poor connectivity: {len(poor_zones)}")
```

### Step 6: Validate Assignment Readiness

```python
gr.validate_assignment()
```

### Step 7: Enhance Connectivity (If Needed)

```python
# Only run if there are poorly connected zones
if len(poor_zones) > 0:
    print(f"Adding connectors for {len(poor_zones)} poorly connected zones...")
    gr.enhance_connectors()
    
    # Replace link file
    import shutil
    shutil.copy('connected_network/link_updated.csv', 'connected_network/link.csv')
    
    # Re-validate
    print("\nRe-validating accessibility...")
    gr.validate_accessibility()
else:
    print("âœ… All zones have good connectivity!")
```
